## Question 3

```
using System; // visual studio syntex

namespace TriangleArea
{
    class Program
    {
        static void Main(string[] args)
        {
            double a = 2;
            double b = 2;
            double c = 2;
            Console.WriteLine("Area of Triangle: " + triangleArea(a, b, c));
        }

        static double triangleArea(double a, double b, double c)
        {
            double s = (a + b + c) / 2;
            double area = Math.Sqrt(s * (s - a) * (s - b) * (s - c));
            return area;
        }
    }
}
```

# # Question 4

```
using System; // visual studio syntex

namespace ArraySeparation
{
    class Program
    {
        static void Main(string[] args)
        {
            Console.WriteLine("Enter the number of elements to be stored in the array:");
            int n = int.Parse(Console.ReadLine());
            int[] arr = new int[n];
            int[] evenArr = new int[n];
            int[] oddArr = new int[n];

            Console.WriteLine("Enter " + n + " elements in the array:");
            for (int i = 0; i < n; i++)
            {
                Console.Write("element - " + i + " : ");
                arr[i] = int.Parse(Console.ReadLine());
            }

            int evenIndex = 0;
            int oddIndex = 0;
            for (int i = 0; i < n; i++)
            {
                if (arr[i] % 2 == 0)
                {
                    evenArr[evenIndex++] = arr[i];
                }
                else
                {
                    oddArr[oddIndex++] = arr[i];
                }
            }

            Console.WriteLine("The Even elements are:");
            for (int i = 0; i < evenIndex; i++)
            {
                Console.Write(evenArr[i] + " ");
            }
            Console.WriteLine();

            Console.WriteLine("The Odd elements are:");
            for (int i = 0; i < oddIndex; i++)
            {
                Console.Write(oddArr[i] + " ");
            }
            Console.WriteLine();

            Console.ReadLine();
        }
    }
}
```
